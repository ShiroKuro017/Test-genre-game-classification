# Fix games.csv secara lokal menggunakan games.json

---

Notebook ini membangun ulang file `games.csv` dari `games.json`
dataset Steam Games (FronkonGames) sehingga kolom kembali sejajar,
terutama kolom `AppID`.

## Import library

```python
import pandas as pd
import json
```

## Baca `games.json` dari folder lokal

```python
json_path = "games.json"

# games.json dari FronkonGames berbentuk dict dengan AppID sebagai key
with open(json_path, "r", encoding="utf-8") as f:
    games_raw = json.load(f)

type(games_raw), len(games_raw)
```

>Penjelasan singkat: `games_raw` biasanya bertipe `dict` dengan key = AppID dan value = data game.

## Konversi dict JSON ke DataFrame

```python
# games_raw: dict {appid_str: {field1:..., field2:...}}
df = pd.DataFrame.from_dict(games_raw, orient="index")

# index saat ini adalah AppID, jadikan kolom eksplisit
df["AppID"] = df.index

# Opsional: sort by AppID
df = df.sort_values("AppID").reset_index(drop=True)

df.head()
df.shape
```

## Pemetaan nama kolom JSON → nama kolom CSV

```python
col_map = {
    "name": "Name",
    "release_date": "Release date",
    "estimated_owners": "Estimated owners",
    "peak_ccu": "Peak CCU",
    "required_age": "Required age",
    "price": "Price",
    "discount": "Discount",
    "dlc_count": "DLC count",
    "about_the_game": "About the game",
    "supported_languages": "Supported languages",
    "full_audio_languages": "Full audio languages",
    "reviews": "Reviews",
    "header_image": "Header image",
    "website": "Website",
    "support_url": "Support url",
    "support_email": "Support email",
    "windows": "Windows",
    "mac": "Mac",
    "linux": "Linux",
    "metacritic_score": "Metacritic score",
    "metacritic_url": "Metacritic url",
    "achievements": "Achievements",
    "recommendations": "Recommendations",
    "notes": "Notes",
    "average_playtime_2weeks": "Average playtime two weeks",
    "median_playtime_2weeks": "Median playtime two weeks",
    "average_playtime_forever": "Average playtime forever",
    "median_playtime_forever": "Median playtime forever",
    "developers": "Developers",
    "publishers": "Publishers",
    "categories": "Categories",
    "genres": "Genres",
    "tags": "Tags",
    "screenshots": "Screenshots",
    "movies": "Movies",
}

# pastikan hanya kolom yang ada di df yang dipakai
base_cols = ["AppID"]

# Hanya ambil kolom yang benar‑benar ada di JSON lokal
available_json_cols = [c for c in col_map.keys() if c in df.columns]

len(available_json_cols), available_json_cols[:10]
```

## Susun DataFrame yang sudah rapi

```python
# Susun urutan kolom: AppID + kolom lain dari JSON
df_clean = df[base_cols + available_json_cols].copy()

# Ganti nama kolom mengikuti layout CSV referensi
df_clean = df_clean.rename(columns=col_map)

df_clean.head()
df_clean.shape
```

## Simpan kembali ke CSV

```python
# Nama file keluaran di folder yang sama
output_path = "games_fixed.csv"

# Simpan tanpa index agar tidak menambah kolom Unnamed
df_clean.to_csv(output_path, index=False)

print("File tersimpan sebagai:", output_path)
```
